### 1.1 Preprocessing Steps
1. Data Parsing and Merging:

    * Parsed the nested JSON structure (root -> _source -> post) to extract the post_id and text. I skipped Hastag because most of them doesn't have it.
    * Merge the text data with data_identification.csv to identify train/test splits and emotion.csv to get the labels using unique id.
2. Data Visualization & Imbalance Check:

    * Before cleaning, I visualized the target variable.  "Joy" dominates the dataset ~ 50% just like described in competition, while emotions like "Disgust" and "Anger" are minority classes.
    
    Note: This why class-weighting later was used.

3. Text Cleaning Pipeline:
 Functions to normalize the text data:

    * Lowercasing: Converted all text to lower case to make "Help" and "help" are treated as the same word.
    * Noise Removal: Removed URLs, user mentions (@user), and hashtags (#topic) using Regular Expressions (Regex).
    * Special Character Removal: Removed numbers and punctuation, keeping only alphabetic characters.
    * Tokenization & Stopwords: Split text into individual words and removed common English stopwords for example "the", "is", "at", using the NLTK library.
    * Lemmatization: Converted words to their root form such as "running" → "run" to reduce vocabulary size.

### 1.2 Feature Engineering Steps
Convert the cleaned text into numerical vectors using TF-IDF (Term Frequency-Inverse Document Frequency).

Why?

TF-IDF weighs words based on their importance. It penalizes words that appear too frequently across all documents like stop words "the", "and", "of" which usually carry less meanin. Then highlights unique words specific to a certain emotion.

####Configuration
* N-Grams (1,2): extract Unigrams (single words) and Bigrams (pairs of consecutive words).Because Bigrams capture context. For example, "not happy" means the opposite of "happy". And unigram model might miss this distinction.
* Max Features (10,000): limit the vocabulary to the top 10,000 most frequent features to prevent the dataset from becoming too sparse and to improve training speed.


### 1.3 Explanation of Your Model

###Random Forest Classifier
  * My intuition is that we may have features to use but Decision might just prone to overfit since the data are text which have so many features. That is why Random Forest is choosed.
  * Ensemble Method: The Random Forest creates multiple decision trees, in my case 100 trees and merges their predictions. The final class is determined by a majority vote. This reduces the risk of overfitting compared to a single Decision Tree.
  * Handling Class Imbalance: We utilized the parameter class_weight='balanced'.
  * Mechanism: model automatically adjusts weights inversely proportional to class frequencies. It penalizes the model heavily for misclassifying minority classes (like "Fear" or "Disgust") and less for the majority class ("Joy").

# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import pandas as pd
import numpy as np
import json
import re
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# ==========================================
# 1. DATA LOADING & PARSING
# ==========================================
print("Loading data...")

# Load JSON Text Data
# The structure is nested: list -> root -> _source -> post -> [text, post_id]
with open('/content/final_posts.json', 'r') as f:
    posts_data = json.load(f)

# Parse JSON into a list of dictionaries
parsed_data = []
for entry in posts_data:
    try:
        post_info = entry['root']['_source']['post']
        parsed_data.append({
            'id': post_info['post_id'],
            'text': post_info['text']
        })
    except KeyError:
        continue

df_text = pd.DataFrame(parsed_data)

# Load CSVs
df_id = pd.read_csv('/content/data_identification.csv')
df_emotion = pd.read_csv('/content/emotion.csv')

# Rename columns to ensure consistency for merging
# format: "id", "test"
df_id.columns = ['id', 'split_type']

# Merge everything into one master dataframe
# 1. Merge text with ID split info
df = pd.merge(df_text, df_id, on='id', how='left')

# 2. Merge with labels
#only training rows will have labels, test will be nan
df = pd.merge(df, df_emotion, on='id', how='left')

print(f"Total rows: {len(df)}")
print(df.head())
```

```python
# 2. EDA & VISUALIZATION

print("\nAnalyzing Class Imbalance...")

# Filter out the training set for visualization
train_df = df[df['split_type'] == 'train'].copy()

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='emotion', data=train_df, order=train_df['emotion'].value_counts().index)
plt.title('Distribution of Emotions in Training Data')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.show()

# Calculate weights/proportions
print(train_df['emotion'].value_counts(normalize=True))
```

```python
#The data are very inbalance, dominated by label 'joy', to make the data to be balance i will just used build
#in function inside the model.
# class_weight='balanced',
```

```python
### Add the code related to the feature engineering steps in cells inside this section
print("\nPreprocessing text...")

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 3. Remove mentions and hashtags
    text = re.sub(r'@\w+|#\w+', '', text)

    # 4. Remove special characters and numbers (keep only letters)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 5. Tokenize, remove stopwords, and lemmatize
    tokens = text.split()
    filtered_tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]

    return " ".join(filtered_tokens)

df['cleaned_text'] = df['text'].apply(clean_text)
```

## 2. Feature Engineering Steps

```python
print("\nExtracting Features (TF-IDF)...")

# Initialize TF-IDF
# ngram_range=(1, 2): Uses both single words and pairs of words (bi-grams)
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))

# Fit on ALL text (train + test) to ensure vocabulary covers everything
X_all_vectors = tfidf.fit_transform(df['cleaned_text'])
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
print("\nPreparing Model...")

# Split into Train and Test
train_mask = df['split_type'] == 'train'
test_mask = df['split_type'] == 'test'

X_train_full = X_all_vectors[train_mask.to_numpy()]
y_train_full = df.loc[train_mask, 'emotion']

X_test_final = X_all_vectors[test_mask.to_numpy()]
id_test_final = df.loc[test_mask, 'id']

# Split Training data and validation (80/20 split)
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.2, random_state=42, stratify=y_train_full
)

# Model: Logistic Regression with 'class_weight="balanced"'
# This automatically assigns higher weights to minority classes (Anger, Disgust)
# and lower weights to majority classes (Joy).
model = LogisticRegression(class_weight='balanced', max_iter=1000, multi_class='ovr')

print("Training Logistic Regression...")
model.fit(X_train, y_train)

# Validation
val_preds = model.predict(X_val)
print("\nValidation Results:")
print(classification_report(y_val, val_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest
# n_estimators=100 , build 100 trees and average their results
model = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1  # Uses all CPU cores to speed up training
)

print("Training Random Forest...")
model.fit(X_train, y_train)

# Validation
val_preds_rf = model.predict(X_val)
print("\nRandom Forest Validation Results:")
print(classification_report(y_val, val_preds_rf))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    class_weight='balanced',
    random_state=42,
    max_depth=50
)

print("Training Decision Tree...")
model.fit(X_train, y_train)

# 4. Validation Report
val_preds = model.predict(X_val)
print("\nDecision Tree Validation Results:")
print(classification_report(y_val, val_preds))
```

```python
print("Retraining on full data and generating submission...")

# Train on all available training data
model.fit(X_train_full, y_train_full)

# Predict on the official test set
final_predictions = model.predict(X_test_final)

# Create submission DataFrame
submission_df = pd.DataFrame({
    'id': id_test_final,
    'emotion': final_predictions
})

# Save to CSV
submission_df.to_csv('submission.csv', index=False)
print("submission.csv created successfully!")

# Check first few rows
print(submission_df.head())
```
